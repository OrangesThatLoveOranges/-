```python
# coding:utf-8

import os
import random
import argparse

parser = argparse.ArgumentParser()
#train文件的地址，根据自己的数据进行修改 xml一般存放在Annotations下
parser.add_argument('--xml_path', default='images//train', type=str, help='input xml label path')
#val文件的地址
parser.add_argument('--txt_path', default='images//val', type=str, help='output txt label path')
opt = parser.parse_known_args()[0]


trainfilepath = opt.xml_path
valsavepath = opt.txt_path
train_img = os.listdir(trainfilepath)
val_img=os.listdir(valsavepath)



file_train = open('train.txt', 'w')


train_index=range(len(train_img))
val_index=range(len(val_img))
num=0
#分配训练集
for i in train_index:
    num+=1
    name = train_img[i][:-4] + '.jpg'
    file_train.writelines("datasets/font/images/train/"+name+"\n")
    if num>3000:
        break

file_train.close()
```

```python
file_val = open('val.txt', 'w')
#分配测试集
for i in val_index:
    name = val_img[i][:-4] + '.jpg'
    file_val.writelines("datasets/font/images/val/"+name+"\n")
file_val.close()
```

```python
#将数据集格式转换为yolo
from PIL import Image
def train():
    for img_name in img_train:
        
        img = Image.open("images//train//" + img_name[0:-1]+".jpg")
        boxs = open("labels//train//" + img_name[0:-1] + ".txt", "rb+").readlines()
        h, w = img.height,img.width
        r_box=""
        for box in boxs:
            box = str(box)
            box = box.split(',')[0:8]
            center_x = ((float(box[6]) + float(box[2])) / 2) / w
            center_y = ((float(box[1]) + float(box[3])) / 2) / h
            height = (abs(float(box[3]) - float(box[1]))) / h
            width = (abs(float(box[6]) - float(box[2]))) / w
            r_box += "0 " + str(center_x) + " " + str(center_y) + " " + str(width) + " " + str(height) + "\n"
        txt_file = open("labels//train//" + img_name[0:-1] + ".txt", "w")
        txt_file.write("")
        txt_file.write(r_box)
        txt_file.close()
       


            
train()
        
```
